```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../../sample_data/Sales.csv')
```

```python
df.head(6)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
random_products = ['Allienware 360hz', 'Logitech Superlight', 'AMD Ryzen 5500', 'RTX 3060']
df['Product'] = df['Product'].apply(lambda x: np.random.choice(random_products) if pd.isna(x) else x)
```

```python
df['OrderAmount'] = df['OrderAmount'].apply(lambda x: np.random.randint(0,4000) if pd.isna(x) else x)
```

```python
df['QuantityOrdered'] = df['QuantityOrdered'].apply(lambda x: np.random.randint(0,20) if pd.isna(x) else x)
```

```python
df['ShippingCost'] = df['ShippingCost'].apply(lambda x: np.random.randint(0,50) if pd.isna(x) else x)
```

```python
status_choice = ['Paid', 'Pending', 'Failed']
df['PaymentStatus'] = df['PaymentStatus'].apply(lambda x: np.random.choice(status_choice) if pd.isna(x) else x)
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format='%Y-%m-%d')
```
