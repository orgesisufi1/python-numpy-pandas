# 1. Creating Dataframe from list and dictionary

```python
import pandas as pd
```

```python
data_list = [
    [1, 'Alice', 25, 'Engineer'],
    [2, 'Bob', 30, 'Doctor'],
    [3, 'Charlie', 35, 'Teacher'],
    [4, 'David', 28, 'Designer']
]
```

```python
data_dict = {
    'ID': [101, 102, 103, 104],
    'Name': ['John', 'Emma', 'Liam', 'Sophia'],
    'Age': [22, 34, 27, 40],
    'Occupation': ['Artist', 'Scientist', 'Writer', 'Chef']
}
```

```python
df1 = pd.DataFrame(data_list)
df2 = pd.DataFrame(data_dict)
```

```python
df1.rename(columns={0:'ID', 1:'Name', 2: 'Age', 3: 'Occupation'}, inplace=True)
```

```python
df1.set_index('ID', inplace=True)
df2.set_index('ID', inplace=True)
```

# Filtering columns

```python
filt_asia = csv_data['Continent'] == 'Asia'
csv_data[filt_asia]
```

```python
filt_2022 = csv_data['2022Population'] > 100000000
csv_data[filt_2022]
```

```python
filt_2000 = csv_data['2000Population'] < 10000000
csv_data[filt_2000]
```

```python
filt_den = csv_data['Density'] > 500
csv_data[filt_den]
```

```python
filt_grw = csv_data['GrowthRate'] < 1
csv_data[filt_grw]
```

```python
sorted_val = csv_data.sort_values(by='2022Population',ascending=False)
sorted_val.head(10)
```

```python
small_filt = csv_data['Area'] < 1000
csv_data[small_filt]
```

```python
filt_capital = ((csv_data['Capital'].str.startswith('B')) & (csv_data['2022Population'] > 20000000))
csv_data[filt_capital]
```

```python
filt_population = csv_data['2022Population'].sum()
filt_countries = (((csv_data['2022Population'] / filt_population)*100)>1)
csv_data[filt_countries]
```

```python
filt_a = (csv_data['Country/Territory'].str.startswith('A'))
new_csv = csv_data[filt_a][['Rank', 'CCA3', 'Country/Territory', 'Capital', 'Continent']]
new_csv.to_csv('datas/orges_world_data.csv')
```

```python
filt_100 = (csv_data['2020Population'] > 100000000)
new_excel = csv_data[filt_100]
new_excel.to_excel('datas/orges_world_data.xlsx')
```
