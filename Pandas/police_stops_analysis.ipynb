## Derivative Columns

```python
df['stop_date'] = pd.to_datetime(df['stop_date'], format='%Y-%m-%d')
```

```python
df['stop_hour'] = pd.to_datetime(df['stop_time'], format='%H:%M').dt.time
```

## Non-speeding Arrests

```python
df['is_weekend'] = ((pd.to_datetime(df['stop_date']).dt.day_name()) == 'Sunday') | ((pd.to_datetime(df['stop_date']).dt.day_name()) == 'Saturday')
```

```python
arr = df[(df['is_arrested']==True) & (~(df['violation']=='Speeding'))]
arr
```

```python
percentage = (arr['stop_date'].count() / df['stop_date'].count())*100
percentage
```

```python
citation = df[((df['stop_outcome']) == 'Citation') & (((df['stop_time']) >= '18:00')&((df['stop_time']) <= '6:00'))]
citation
```

```python
violation_grp = df.groupby(['violation'])
```

```python
drv_mean = violation_grp['driver_age'].mean()
stop_count = violation_grp['stop_date'].count()
```

```python
grouping = pd.DataFrame({'average_age': drv_mean, 'total_stops': stop_count})
grouping
```
