## Filtering songs

```python
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
```

```python
more_than = df['streams'] > 10000000
more_than
```

```python
artist_counts = df['artist_count'] >= 3
artist_counts
```

```python
summer = ((df['released_year']==2023) & ((df['released_month']>=6)&(df['released_day']>=21)) | ((df['released_month']>=7)&(df['released_month']<=8)))
summer.head(10)
```

```python
df.info()
```

```python
top10 = df.sort_values(by='streams', ascending=False)
spotify_pls = (top10['in_apple_playlists']>top10['in_spotify_charts'])
spotify_pls
```

## Calculating columns

```python
def to_date(row):
    year, month, day = row['released_year'], row['released_month'], row['released_day']
    date = f'{year}-{month}-{day}'
    return pd.to_datetime(date)

df['release_date'] = df[['released_year', 'released_month', 'released_day']].apply(to_date, axis=1)
df.head()
```

```python
def calculate_stream_percentage(df):
    max_streams = df['streams'].max() 
    df['stream_percentage'] = (df['streams'] / max_streams) * 100
    return df


df = calculate_stream_percentage(df)
df.head()
```

```python
def perc_playlist(df):
    total = df['in_spotify_playlists']+df['in_apple_playlists']
    df['spotify_percentage'] = (df['in_spotify_playlists']/total)*100
    return df

df = perc_playlist(df)
df.head()
```

## Grouping

```python
copied_spotify = df.copy()
copied_spotify['artist(s)_name'] = copied_spotify['artist(s)_name'].apply(lambda x: x.split(',')[0])
copied_spotify['artist(s)_name']
```

```python
df.head()
```

```python
artist_grp = copied_spotify.groupby('artist(s)_name')
```

```python
artist_grp['streams'].apply(lambda x: x.sort_values(ascending=False)).sort_values(ascending=False)
```

```python
split_artists = df['artist(s)_name'].str.split(',')
all_aritsts = [name.strip() for record in split_artists for name in record]
artists = pd.Series(all_aritsts).value_counts()
artists
```

# Creating a report file

```python
songs = df['track_name']
report_df = pd.DataFrame({'track_name': songs,'more_than_10M_streams':more_than, 'three_or_more_artist':artist_counts, 'Summer':summer, 'top10_more_in_spotify_than_apple':spotify_pls})
report_df
```

```python
merged_df = df.merge(report_df, how='left', on='track_name')
merged_df
```

```python
merged_df.to_csv('datas/spotify_data.csv', index=False)
```
